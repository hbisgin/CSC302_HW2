```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('car_crashes')
print(df.shape)
df.head()
```

```python
sns.scatterplot(data=df, x='alcohol', y='speeding', size='ins_premium', marker='o' )
plt.xlabel('alchol consumption')
plt.legend(title="insurance premium", loc=4) #location can be given as a string too, i.e., lower right
plt.title("Alcohol Consumption vs. Speeding in US states")
```

```python
sns.regplot(data=df, x='alcohol', y='speeding')
```

```python
df = sns.load_dataset('penguins')
df.dropna(inplace=True)
print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
sns.boxplot(data=df, x='sex', y='bill_depth_mm')
plt.ylabel('biil depth (mm)')
plt.xlabel('')
```

```python
species = sns.relplot(data=df, x='bill_length_mm', y='bill_depth_mm', col="species", hue="sex", kind='scatter', size='body_mass_g')
titles=['Adelie', 'Chinstrap', 'Gentoo']
for ax,title in zip(species.axes.flatten(),titles):
    ax.set_title(title )
    ax.set_xlabel(title + 'bill length (mm)')

species.set_ylabels('bill depth (mm)')
species.add_legend(loc=4)

```
